```python
import numpy as np
import pandas as pd
```

# Loading Data

```python
df = pd.read_csv('creditcard.csv')
df
```

# Data Preprocessing 

```python
# check info

df.info()
```

```python
df.describe()
```

```python
df.value_counts()
```

```python
# Check null values

df.isna().sum()
```

```python
# Check Duplicated Values

print(df.duplicated().sum())

# Remove duplicated Values

df.drop_duplicates(inplace=True)

print(df.duplicated().sum())
```

# Splitting Data

```python
# Splitting the data into testing and training

x = df.drop('Class', axis=1)
y = df['Class']

print(x.shape)
print(y.shape)

from sklearn.model_selection import train_test_split

x_train, x_test,y_train, y_test = train_test_split(x, y , test_size = 0.20)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train
```

```python
y_train
```

# Apply Standard Scaler

```python
# Apply Standard Scaler on the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# scaler.fit(x_train)

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)


print(x_train.shape)
print()
print(x_test.shape)
print()
print(x_train)
print()
print(x_test)
print()
```

```python
x_train
```

# Apply SMOTE to Handle Class Imbalance

```python
# Check Class Imbalance

y_train.value_counts()
```

```python
# Apply SMOTE Analysis

from imblearn.over_sampling import SMOTE

smote = SMOTE()

x_train,y_train = smote.fit_resample(x_train,y_train)

print(x_train.shape)
print()
print(y_train.shape)
print()
```

```python
y_train.value_counts()
```

# Apply PCA for Dimentionality Reduction

```python
# Apply PCA for Dimentionality Reduction

from sklearn.decomposition import PCA

pca = PCA(n_components=0.99)

pca.fit(x_train)

x_train = pca.transform(x_train)
x_test = pca.transform(x_test)

print(x_train.shape)
print(x_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()

model_lr.fit(x_train,y_train)
```

```python
# model training score

model_lr.score(x_train,y_train)
```

```python
# model training score

model_lr.score(x_test,y_test)
```

```python
# Model Prediction

model_predicted = model_lr.predict(x_test)
model_predicted
```

```python
# Model Accuracy Score

from sklearn.metrics import accuracy_score

accuracy_score(y_test,model_predicted)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model_dtc = DecisionTreeClassifier()

model_dtc.fit(x_train,y_train)
```

```python
# model training score

model_dtc.score(x_train,y_train)
```

```python
# model testing score

model_dtc.score(x_test,y_test)
```

```python
# mdoel prediction

model_predicted_dtc = model_dtc.predict(x_test)
model_predicted_dtc
```

```python
# Model Accuracy Score

from sklearn.metrics import accuracy_score

accuracy_score(y_test,model_predicted_dtc)
```
