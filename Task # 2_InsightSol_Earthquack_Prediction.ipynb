# After Correlation

```python
# correlation from 2 to onward
df = df[['cdi','nst','dmin','gap','country','year','alert_green','alert_nan','magType_mwb',
        'magType_mwc','magType_mww','tsunami']]
df.shape
```

```python
# # correlation near above and equal to 0.4

# df = df[['nst','dmin','year','alert_green','alert_nan',
#         'magType_mwc','magType_mww','tsunami']]
```

```python
x = df.drop('tsunami', axis=1)
x.head(5)
```

```python
y = df['tsunami']
y.head()
```

```python
# Split Data into training and testing

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.20)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# Apply Standarad Scaler

```python
# Apply Standarad Scaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

print(x_train.shape)
print(x_test.shape)
```

# Check Class Imbalance _Apply SMOTE Analysis

```python
df.tsunami.value_counts()
```

```python
# Apply SMOTE Analysis

from imblearn.over_sampling import SMOTE

smote = SMOTE()

x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
print(y_train.value_counts())
```

```python
x_train.shape
```

# Apply PCA for dimentionality reduction

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.99)

pca.fit(x_train)

x_train = pca.transform(x_train)
x_test = pca.transform(x_test)
```

```python
print(x_train.shape)
print(x_test.shape)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model_rfc = RandomForestClassifier()

model_rfc
```

```python
model_rfc.fit(x_train,y_train)
```

```python
# model training score

model_rfc.score(x_train,y_train)
```

```python
# model testing score

model_rfc.score(x_test,y_test)
```

```python
model_predicted = model_rfc.predict(x_test)
```

```python
# model accuracy score

from sklearn.metrics import accuracy_score

accuracy_score(y_test,model_predicted)
```

# Apply KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=8)
knn
```

```python
knn.fit(x_train,y_train)
```

```python
# model training score

knn.score(x_train, y_train)
```

```python
# model testing score

knn.score(x_test, y_test)
```

```python
# MODEL PREDICTED VALUES

knn_predicted = knn.predict(x_test)
knn_predicted
```

```python
# model accuracy score

accuracy_score(knn_predicted, y_test)
```
